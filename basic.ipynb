## Resizing and Rescaling

```python
#Resize image and plotting original and resized image
img = cv2.imread('images/flower.jpg') # read image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
resized_img = cv2.resize(img, (100, 100)) # resize image
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.imshow(img) # show image
plt.title('Original image') # set title
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.imshow(resized_img) # show image
plt.title('Resized image') # set title
plt.show() # show plot
```

```python
# Rescale image
img = cv2.imread('images/flower.jpg') # read image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
scale_percent = 60 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
rescaled_img = cv2.resize(img, dim) # resize image
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.imshow(img) # show image
plt.title('Original image') # set title
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.imshow(rescaled_img) # show image
plt.title('Rescaled image') # set title
plt.show() # show plot
```

```python
#drawing line
img = cv2.imread('images/flower.jpg') # read image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
cv2.line(img, (0,0), (img.shape[1], img.shape[0]), (255, 0, 0), 20) # draw a line
plt.imshow(img) # show image
plt.title('Line') # set title
plt.show() # show plot
```

```python
#drawing rectangle
img = cv2.imread('images/flower.jpg') # read image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
cv2.rectangle(img, (1000,1000), (250, 350), (0, 255, 0), 5) # draw a rectangle
# (1000,1000) --> top left corner
# (250, 350) --> bottom right corner
# (0, 255, 0) --> color
# 5 --> thickness
plt.imshow(img) # show image
plt.title('Rectangle') # set title
plt.show() # show plot
```

```python
#drawing filled rectangle
img = cv2.imread('images/flower.jpg') # read image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
cv2.rectangle(img, (1000,1000), (250, 350), (0, 255, 0), cv2.FILLED) # draw a filled rectangle
# (1000,1000) --> top left corner
# (250, 350) --> bottom right corner
# (0, 255, 0) --> color
# cv2.FILLED --> thickness
plt.imshow(img) # show image
plt.title('Filled rectangle') # set title
plt.show() # show plot
```

```python
#drwaing circle
img = cv2.imread('images/flower.jpg') # read image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
cv2.circle(img, (400, 500), 100, (255, 0, 0), 30) # draw a circle
# (400, 500) --> center
# 100 --> radius
# (255, 0, 0) --> color
# 30 --> thickness
plt.imshow(img) # show image
plt.title('Circle') # set title
plt.show() # show plot
```

```python
#drawing text
img = cv2.imread('images/flower.jpg') # read image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
cv2.putText(img, 'OpenCV', (500, 500), cv2.FONT_HERSHEY_COMPLEX, 5, (255, 255, 255), 2) # draw text
# 'OpenCV' --> text
# (500, 500) --> bottom left corner
# cv2.FONT_HERSHEY_COMPLEX --> font
# 1 --> scale
# (0, 150, 0) --> color
# 2 --> thickness
plt.imshow(img) # show image
plt.title('Text') # set title
plt.show() # show plot
```
